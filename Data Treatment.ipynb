# Data Cleaning

```python
import numpy as np
import pandas as pd
```

```python
def impute(value):
    if value==np.inf:
        return 1
    elif value==-np.inf:
        return -1
    elif value is np.NaN:
        return 0
    else: 
        return np.round(np.sqrt(value),2)
```

```python
df=pd.read_csv("input_data.csv")
df["dataPoints"]=df["dataPoints"].apply(impute)
df1=df.replace(to_replace=np.nan, value=0)
```

```python
df1.to_csv("output.csv",index=False)
```
