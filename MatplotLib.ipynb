## Normal distribution 

```python
mu=100
sg=15
val=np.random.normal(mu,sg,size=1000000)
```

```python
plt.hist(val,bins=100)
plt.show()
```

Most of the values are close to 100, mean=100
70% of the data is present under 1 standard deviatiom (100-15, 100+15)
98% of the data is present under 2 standard deviations (70, 130)
99.9% data lies under 3 std, (55, 145)

## Sigmoid function

```python
# sg=1/1+eˆ-x

# x=(-100,100)
```

```python
lst=np.arange(-100,100)
x=list(map(lambda x:1/(1+ math.exp(-x)),lst))
```

```python
plt.plot(x)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
sig=np.vectorize(sigmoid)

x=np.arange(-100,100)
y=sig(x)
```

```python
plt.plot(x,y)
plt.xlim(-15,15)
plt.ylim(-1,1.5)
plt.show()
```

# Subplot

```python
x=np.arange(0,20,0.1)
sin=np.sin(x)
cos=np.cos(x)
```

```python
plt.plot(x,sin) 

plt.show()
```

```python
x=np.arange(0,20,0.1)
sin=np.sin(x)
cos=np.cos(x)
tan=np.tan(x)

plt.figure()

plt.subplot(2,3,1)
plt.plot(x,sin)

plt.subplot(2,3,3)
plt.plot(x,cos)

plt.subplot(2,3,5)
plt.plot(x,tan)

plt.show()
```

```python
fig = plt.figure()
ax231 = fig.add_subplot(231)
ax231.plot(x, sin)
ax233 = fig.add_subplot(233)
ax233.plot(x, tan)
ax235 = fig.add_subplot(235)
ax235.plot(x, cos)
ax231.set_title("TITLE 1")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
np.random.seed(0) 
df = pd.DataFrame(data={'1':np.random.randint(0, 100, 10), '2':np.random.randint(0, 100, 10)})
plt.bar(df.index.values, df['1'])
plt.bar(df.index.values,df['2'],bottom=df['1'])

plt.legend(['1', '2'])
plt.show()
```
